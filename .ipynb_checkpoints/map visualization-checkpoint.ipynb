```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

```python
df_crime = pd.read_csv("https://cocl.us/sanfran_crime_dataset") 
# let's rename the columns so that they make sense
df_crime.rename(columns={'PdDistrict':'Neighborhood', 'Resolution':'Count'}, inplace=True)
df_crime.head()
```

```python
print(df_crime.shape)
```

```python
df=df_crime.groupby('Neighborhood').size().reset_index(name='Count')
df
```

```python

world_geo =  r'https://cocl.us/sanfran_geojson/san-francisco.geo.json' # geojson file
# create a plain world map
world_map = folium.Map(location=[37.75, -122.44], zoom_start=11)
```

```python
# generate choropleth map using the total immigration of each country to Canada from 1980 to 2013
world_map.choropleth(
    geo_data=world_geo,
    data=df,
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco'
)

# display map
world_map
```
